```python
import torch
from torch.utils.data import Dataset

class StudentSequenceDataset(Dataset):
    def __init__(
        self,
        df,
        feature_cols,
        max_len=32,
        id_col="student_id"
    ):
        self.df = df.copy()
        self.feature_cols = feature_cols
        self.max_len = max_len
        self.id_col = id_col

        # Keep only rows with labels
        self.df = self.df.dropna(subset=["y_next_score", "y_next_difficulty", "y_improved"])

        # index by student and position
        self.seq_index = self._build_index()

    def _build_index(self):
        seq_index = []
        grouped = self.df.groupby(self.id_col)
        for sid, g in grouped:
            idxs = g.index.to_numpy()
            if len(idxs) < 2:
                continue
            for i in range(len(idxs)):
                end = i + 1
                start = max(0, end - self.max_len)
                seq_idx = idxs[start:end]
                target_idx = idxs[end - 1]
                seq_index.append((seq_idx, target_idx))
        return seq_index

    def __len__(self):
        return len(self.seq_index)

    def __getitem__(self, idx):
        seq_idx, target_idx = self.seq_index[idx]

        seq_df = self.df.loc[seq_idx]
        x = torch.tensor(seq_df[self.feature_cols].values, dtype=torch.float32)

        # padding
        if len(seq_df) < self.max_len:
            pad_len = self.max_len - len(seq_df)
            pad = torch.zeros(pad_len, x.shape[1])
            x = torch.cat([pad, x], dim=0)

        y_score = torch.tensor(self.df.loc[target_idx, "y_next_score"], dtype=torch.float32)
        y_diff = torch.tensor(self.df.loc[target_idx, "y_next_difficulty"], dtype=torch.float32)
        y_improved = torch.tensor(self.df.loc[target_idx, "y_improved"], dtype=torch.float32)

        return {
            "x": x,  # [max_len, n_features]
            "y_score": y_score,
            "y_diff": y_diff,
            "y_improved": y_improved,
        }
```

```python
import math
import torch
import torch.nn as nn
import pytorch_lightning as pl
from torchmetrics import MeanSquaredError, AUROC

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=512):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len).unsqueeze(1).float()
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer("pe", pe.unsqueeze(0))  # [1, max_len, d_model]

    def forward(self, x):
        # x: [batch, seq_len, d_model]
        seq_len = x.size(1)
        return x + self.pe[:, :seq_len, :]


class StudentTransformerModel(pl.LightningModule):
    def __init__(
        self,
        d_in,
        d_model=64,
        nhead=4,
        num_layers=2,
        dim_feedforward=128,
        dropout=0.1,
        lr=1e-3,
        max_len=128,
        pos_encoding=True,
    ):
        super().__init__()
        self.save_hyperparameters()

        self.input_proj = nn.Linear(d_in, d_model)
        self.pos_encoder = PositionalEncoding(d_model, max_len) if pos_encoding else nn.Identity()

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=dim_feedforward,
            dropout=dropout,
            batch_first=True,
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        self.score_head = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Linear(d_model, 1),
        )
        self.diff_head = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Linear(d_model, 1),
        )
        self.improve_head = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Linear(d_model, 1),
        )

        self.mse_score = MeanSquaredError()
        self.mse_diff = MeanSquaredError()
        self.auroc = AUROC(task="binary")

        self.lr = lr

    def forward(self, x):
        # x: [batch, seq_len, d_in]
        h = self.input_proj(x)
        h = self.pos_encoder(h)
        h = self.encoder(h)  # [batch, seq_len, d_model]
        h_last = h[:, -1, :]  # use last token

        y_score = self.score_head(h_last).squeeze(-1)
        y_diff = self.diff_head(h_last).squeeze(-1)
        y_improve_logit = self.improve_head(h_last).squeeze(-1)

        return y_score, y_diff, y_improve_logit

    def _shared_step(self, batch, stage):
        x = batch["x"]
        y_score_true = batch["y_score"]
        y_diff_true = batch["y_diff"]
        y_improved_true = batch["y_improved"]

        y_score_pred, y_diff_pred, y_imp_logit = self(x)

        # losses
        mse_score = nn.functional.mse_loss(y_score_pred, y_score_true)
        mse_diff = nn.functional.mse_loss(y_diff_pred, y_diff_true)

        mask = ~torch.isnan(y_improved_true)
        if mask.sum() > 0:
            bce = nn.functional.binary_cross_entropy_with_logits(
                y_imp_logit[mask],
                y_improved_true[mask]
            )
        else:
            bce = torch.tensor(0., device=self.device)

        loss = mse_score + mse_diff + bce

        # logging
        self.log(f"{stage}_loss", loss, prog_bar=True)
        self.log(f"{stage}_mse_score", mse_score, prog_bar=False)
        self.log(f"{stage}_mse_diff", mse_diff, prog_bar=False)

        if mask.sum() > 0:
            probs = torch.sigmoid(y_imp_logit[mask])
            self.auroc.update(probs, y_improved_true[mask].int())
            self.log(f"{stage}_bce", bce, prog_bar=False)

        return loss

    def training_step(self, batch, batch_idx):
        return self._shared_step(batch, "train")

    def validation_step(self, batch, batch_idx):
        self._shared_step(batch, "val")

    def on_validation_epoch_end(self):
        try:
            auroc = self.auroc.compute()
            self.log("val_auroc", auroc, prog_bar=True)
        except:
            pass
        self.auroc.reset()

    def configure_optimizers(self):
        opt = torch.optim.AdamW(self.parameters(), lr=self.lr)
        sched = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="min", patience=3)
        return {
            "optimizer": opt,
            "lr_scheduler": {
                "scheduler": sched,
                "monitor": "val_loss",
            },
        }
```

# Preparing train and validation data

```python
from torch.utils.data import DataLoader, random_split
import pytorch_lightning as pl

# Suppose df_fe is the feature-engineered dataframe
feature_cols = [
    "time_since_prev",
    "cum_interactions",
    "cum_score_mean",
    "score_per_difficulty",
    "roll_mean_score_3",
    "roll_mean_score_5",
    "roll_mean_score_10",
    "roll_sum_time_spent_3",
    "roll_sum_time_spent_5",
    "roll_mean_wait_3",
    "roll_mean_wait_5",
    "difficulty",
    "wait_time",
    "resource_score",
    "time_spent",
]


FEATURES_TO_SCALE = [
    "time_since_prev",
    "cum_interactions",
    "cum_score_mean",
    "score_per_difficulty",
    "roll_mean_score_3",
    "roll_mean_score_5",
    "roll_mean_score_10",
    "roll_sum_time_spent_3",
    "roll_sum_time_spent_5",
    "roll_mean_wait_3",
    "roll_mean_wait_5",
    "wait_time",
    "resource_score",
    "time_spent",
]


from sklearn.preprocessing import RobustScaler

student_ids = df_fe_clean["student_id"].unique()
rng = np.random.default_rng(42)

train_students = rng.choice(
    student_ids,
    size=int(0.8 * len(student_ids)),
    replace=False
)

train_df = df_fe_clean[df_fe_clean["student_id"].isin(train_students)]
val_df = df_fe_clean[~df_fe_clean["student_id"].isin(train_students)]

scaler = RobustScaler()
train_df[FEATURES_TO_SCALE] = scaler.fit_transform(train_df[FEATURES_TO_SCALE])
val_df[FEATURES_TO_SCALE] = scaler.transform(val_df[FEATURES_TO_SCALE])

train_ds = StudentSequenceDataset(train_df, feature_cols, max_len=32)
val_ds = StudentSequenceDataset(val_df, feature_cols, max_len=32)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True, num_workers=4)
val_loader = DataLoader(val_ds, batch_size=64, shuffle=False, num_workers=4)

model = StudentTransformerModel(d_in=len(feature_cols), d_model=64, num_layers=2, nhead=4)
```

# Training

```python
from pytorch_lightning.loggers import CSVLogger

logger = CSVLogger("logs", name="student_transformer_model")

trainer = pl.Trainer(
    max_epochs=30,
    accelerator="gpu" if torch.cuda.is_available() else "cpu",
    precision="16-mixed" if torch.cuda.is_available() else "32",
    logger=logger,
)

trainer.fit(model, train_loader, val_loader)
```

# Display training and validation metrics

```python
import pandas as pd
import os

# Define the path to the logs directory
log_dir = "logs/student_transformer_model"

# Find the latest version directory
versions = [d for d in os.listdir(log_dir) if d.startswith('version_')]
if not versions:
    raise FileNotFoundError(f"No version directories found in {log_dir}")

latest_version = sorted(versions, key=lambda x: int(x.split('_')[1]))[-1]
metrics_path = os.path.join(log_dir, latest_version, "metrics.csv")

# Load the metrics CSV
if not os.path.exists(metrics_path):
    raise FileNotFoundError(f"Metrics file not found at {metrics_path}")

metrics_df = pd.read_csv(metrics_path)

print(f"Metrics loaded from: {metrics_path}")
display(metrics_df.head())
```

```python
import matplotlib.pyplot as plt

# Ensure epoch is treated as an integer for grouping
metrics_df['epoch'] = metrics_df['epoch'].ffill().astype(int)

# Aggregate training loss: take the mean train_loss for each epoch
train_loss_per_epoch = metrics_df.dropna(subset=['train_loss']).groupby('epoch')['train_loss'].mean().reset_index()

# Aggregate validation loss: take the last recorded val_loss for each epoch
# This handles NaNs and ensures one point per epoch
val_loss_per_epoch = metrics_df.dropna(subset=['val_loss']).groupby('epoch')['val_loss'].last().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(train_loss_per_epoch['epoch'], train_loss_per_epoch['train_loss'], label='Train Loss (Epoch Avg)')
plt.plot(val_loss_per_epoch['epoch'], val_loss_per_epoch['val_loss'], label='Validation Loss (Epoch End)')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ensure epoch is treated as an integer for grouping
metrics_df['epoch'] = metrics_df['epoch'].ffill().astype(int)

# Aggregate training MSE score: take the mean train_mse_score for each epoch
train_mse_score_per_epoch = metrics_df.dropna(subset=['train_mse_score']).groupby('epoch')['train_mse_score'].mean().reset_index()

# Aggregate validation MSE score: take the last recorded val_mse_score for each epoch
val_mse_score_per_epoch = metrics_df.dropna(subset=['val_mse_score']).groupby('epoch')['val_mse_score'].last().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(train_mse_score_per_epoch['epoch'], train_mse_score_per_epoch['train_mse_score'], label='Train MSE Score (Epoch Avg)')
plt.plot(val_mse_score_per_epoch['epoch'], val_mse_score_per_epoch['val_mse_score'], label='Validation MSE Score (Epoch End)')
plt.title('Training and Validation MSE for Score over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE Score')
plt.legend()
plt.grid(True)
plt.show()
```

# Model Inference

```python
import os
import glob

# Define the path to the logger directory
log_dir = "logs/student_transformer_model"

# Find the latest version directory (e.g., version_0, version_1, etc.)
versions = [d for d in os.listdir(log_dir) if d.startswith('version_')]
if not versions:
    raise FileNotFoundError(f"No version directories found in {log_dir}")

latest_version = sorted(versions, key=lambda x: int(x.split('_')[1]))[-1]
checkpoint_dir = os.path.join(log_dir, latest_version, "checkpoints")

# Find the latest checkpoint file within that directory
checkpoint_files = glob.glob(os.path.join(checkpoint_dir, "*.ckpt"))
if not checkpoint_files:
    raise FileNotFoundError(f"No checkpoint files found in {checkpoint_dir}")

# Get the most recently modified checkpoint file
latest_checkpoint_path = max(checkpoint_files, key=os.path.getmtime)

print(f"Latest checkpoint found: {latest_checkpoint_path}")

# Load the model from the checkpoint
# Ensure StudentTransformerModel and other necessary classes are defined in the notebook
loaded_model = StudentTransformerModel.load_from_checkpoint(latest_checkpoint_path)

# Set the model to evaluation mode
loaded_model.eval()

print("Model loaded successfully and set to evaluation mode.")
```

```python
import torch

# Define the path to save the model
model_save_path = "student_transformer_model.pth"

# Save the model's state_dict
torch.save(loaded_model.state_dict(), model_save_path)

print(f"Model saved successfully to {model_save_path}")
```

```python
import torch

# Get a sample batch from the validation loader
# Ensure the data loader is iterable
sample_batch = next(iter(val_loader))

# Move batch to the same device as the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
loaded_model.to(device)

x_sample = sample_batch["x"].to(device)

# Perform inference
with torch.no_grad(): # Disable gradient calculation for inference
    y_score_pred, y_diff_pred, y_improved_logit_pred = loaded_model(x_sample)

# Convert predictions to CPU and numpy for easier viewing
y_score_pred_np = y_score_pred.cpu().numpy()
y_diff_pred_np = y_diff_pred.cpu().numpy()
y_improved_prob_pred_np = torch.sigmoid(y_improved_logit_pred).cpu().numpy()

print("Sample Predictions:")
print("Predicted Next Score (first 5):", y_score_pred_np[:10])
print("Predicted Next Difficulty (first 5):", y_diff_pred_np[:10])
print("Predicted Improved Probability (first 5):", y_improved_prob_pred_np[:10])

# Optionally, compare with true values if available in the batch
y_score_true = sample_batch["y_score"].cpu().numpy()
y_diff_true = sample_batch["y_diff"].cpu().numpy()
y_improved_true = sample_batch["y_improved"].cpu().numpy()

print("\nTrue Values (first 5):")
print("True Next Score (first 5):", y_score_true[:10])
print("True Next Difficulty (first 5):", y_diff_true[:10])
print("True Improved (first 5):", y_improved_true[:10])
```
